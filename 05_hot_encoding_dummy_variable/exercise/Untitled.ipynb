```python
# Predict price of a mercendez benz that is 4 yr old with mileage 45000
# Predict price of a BMW X5 that is 7 yr old with mileage 86000
# Tell me the score (accuracy) of your model.(Hint: use LinearRegression().score())
```

```python
import pandas as pd
```

```python
df = pd.read_csv('carprices.csv')
df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
model = LinearRegression()
le = LabelEncoder()
```

```python
dfle = df
dfle["Car Model"] = le.fit_transform(dfle["Car Model"])
dfle
```

```python
X = dfle[["Car Model","Mileage","Age(yrs)"]].values
X
```

```python
y = dfle["Sell Price($)"].values
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([("Car Model", OneHotEncoder(), [0])], remainder="passthrough")
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:, 2:]
X
```

```python
model.fit(X,y)
```

```python
model.predict([[1, 45000, 4]])
```

```python
model.predict([[0, 86000, 7]])
```

```python
model.score(X,y)
```
